### Automation

```python
def get_heatmap(cuisine):
    cuisine_df = new_data_copy[new_data_copy['Cuisines'].str.contains(cuisine)]
    try:
        basemap = folium.Map()
        HeatMap(cuisine_df[['Latitude', 'Longitude']]).add_to(basemap)
        return basemap
    except ValueError:
        cuisine_df = cuisine_df.isnull().dropna(inplace=True)
        cuisine_df.is_null().sum()
```

```python
get_heatmap('South Indian')
```
